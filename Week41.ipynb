### Gradient descent with fixed learning rate

```python
learning_rate = 1e-1
first_guess = 0
approximations = [first_guess]
tol = 1e-15
while np.abs(df(approximations[-1])) > tol:
    x = approximations[-1]
    approximations.append(x - learning_rate*df(x))
```

```python
x = np.arange(-2,0,0.01)
plt.plot(x, f(x), 'r')
plt.plot(approximations, f(np.asarray(approximations)), 'o')
plt.title("Simple gradient descent")
plt.text(-0.65, -0.45, f"{len(approximations)-1} steps to converge")
```

### Gradient descent with momentum

```python
gamma = 0.2
eta = 0.3
first_guess = 0
approximations = [first_guess]
momentum = [0]
tol = 1e-15
while np.abs(df(approximations[-1])) > tol:
    x = approximations[-1]
    momentum.append(gamma*momentum[-1] + eta*df(x))
    approximations.append(x - momentum[-1])
    
```

```python
x = np.arange(-2,0,0.01)
plt.plot(x, f(x), 'r')
plt.plot(approximations, f(np.asarray(approximations)), 'o')
plt.title("Momentum gradient descent")
plt.text(-0.65, -0.45, f"{len(approximations)-1} steps to converge")
```
